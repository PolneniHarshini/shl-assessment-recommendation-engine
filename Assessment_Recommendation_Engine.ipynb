```python
import pandas as pd

df = pd.read_excel("Gen_AI Dataset.xlsx")
df.head()
```

```python
df.columns
```

```python
df = df.fillna("")

df["combined_text"] = (
    df.astype(str).agg(" ".join, axis=1)
)

df["combined_text"].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words="english")

tfidf_matrix = vectorizer.fit_transform(df["combined_text"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_assessment(user_query, top_n=3):
    user_vector = vectorizer.transform([user_query])
    similarity_scores = cosine_similarity(user_vector, tfidf_matrix)

    top_indices = similarity_scores[0].argsort()[-top_n:][::-1]

    return df.iloc[top_indices]
```

```python
query = "I am hiring a Java developer with good communication skills. Test duration around 40 minutes."
result = recommend_assessment(query)

result
```
